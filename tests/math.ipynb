```python
np.random.seed(42)
vol = 0.2
time = 1
dailyVol = vol / np.sqrt( 252. )
returns = np.random.normal( 0, dailyVol, time * 252 )
print(returns)
priceAtMaturity = np.prod( 1 + returns )
print(priceAtMaturity)
```
